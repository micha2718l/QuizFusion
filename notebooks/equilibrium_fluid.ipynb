```python
from testifyr import (
    WorkProblem,
    Question,
    Test,
    MultipleChoiceProblem,
    Choice,
    Bonus,
    Information,
)
from stuff import Value, Unit
import numpy as np
from pprint import pprint as pp
import base64

problems = []
multiple_choices = []
bonuses = []
g = Value(9.81, "m/s^2")
mode = "dev1"
```

```python
mc = MultipleChoiceProblem(
    statement="What is the relationship between mass, volume, and density?",
    choices=[
        Choice("Density is inversely proportional to both mass and volume."),
        Choice("Density is proportional to both mass and volume."),
        Choice(
            "Density is proportional to mass and inversely proportional to volume.",
            True,
        ),
        Choice("Density is inversely proportional to mass and proportional to volume."),
    ],
)
multiple_choices.append(mc)
if mode == "dev":
    mc.show()
```

```python
depth_a = Value(2, "in")
depth_b = Value(1.5, "ft")

mc = MultipleChoiceProblem(
    statement=f"Zaphod is holding two copies of the Hitchhiker's Guide to the Galaxy (same volume, same density) in a pool of liquid gallium. One of the guides (A) is held {depth_a} below the surface and the other (B) is held at a depth of {depth_b}. Which one experiences a greater buoyant force?",
    choices=[
        Choice("Both feel the same force.", True),
        Choice("Guide A."),
        Choice("Guide B."),
        Choice("Need more information."),
    ],
)
multiple_choices.append(mc)
if mode == "dev":
    mc.show()
# mc.html()
```

```python
rho_water = Value(1000, "kg/m^3")
rho_gallium = Value(6095, "kg/m^3")


def rho(var="", v=None):
    return f"$\\rho_\\mathrm{{{var}}} = {v:C}$"


mc = MultipleChoiceProblem(
    statement=f"Now A is held {depth_a} below the surface in liquid gallium ({rho('gallium', rho_gallium)}) and B is held {depth_b} under water ({rho('water', rho_water)}), which one experiences a greater buoyant force?",
    choices=[
        Choice("Both feel the same force."),
        Choice("Guide A.", True),
        Choice("Guide B."),
        Choice("Need more information."),
    ],
)
multiple_choices.append(mc)
if mode == "dev":
    mc.show()
```

```python
rho_guide = Value(2183, "kg/m^3")

mc = MultipleChoiceProblem(
    statement=f"In the case of Guide B (({rho('Guide', rho_guide)}), $\\mathrm{{Depth}}={depth_b:C}$), will it float up in either fluid?",
    choices=[
        Choice("Both."),
        Choice("Only gallium.", True),
        Choice("Only water."),
        Choice("Neither"),
    ],
)
multiple_choices.append(mc)
if mode == "dev":
    mc.show()
```

```python
r_1 = Value(1, "cm")
r_2 = Value(5, "cm")
r_3 = Value(5, "cm")
r_4 = Value(7, "cm")
rho_chocolate_milk = Value(1050, "kg/m^3")
rho_maple_syrup = Value(1370, "kg/m^3")
rho_liquid_soap = Value(1010, "kg/m^3")
rho_fresh_orange_juice = Value(1045, "kg/m^3")
rho_vegetable_oil = Value(915, "kg/m^3")
height_sections = Value(3, "cm")
speed_1 = Value(137, "m/s")
pressure_2 = Value(4.20, "Pa")

statement = f"The figure below shows a situation where a heterogeneous incompressible fluid is flowing through a pipe. The radii of the sections are $r_1={r_1:C}$, $r_2={r_2:C}$, $r_3={r_3:C}$, $r_4={r_4:C}$. Sections 3 and 4 are {height_sections} below sections 1 and 2.\nChoose one of the following to be the fluid in this problem: (Chocolate milk: {rho("c", rho_chocolate_milk)}), (Maple Syrup: {rho("m", rho_maple_syrup)}), (Liquid Soap: {rho("s", rho_liquid_soap)}), (Fresh Orange Juice: {rho("o", rho_fresh_orange_juice)}), (Vegetable Oil: {rho("v", rho_vegetable_oil)})"

questions = [
    Question(
        "Write down the relevant Bernoulli equation that relates sections 2 and and 4. Keep everything as variables and solve for the pressure in section 4.",
        space=9,
    ),
    Question(
        f"If the speed of the fluid in section 1 is {speed_1}, what is the speed in section 2?",
        space=9,
    ),
    Question(
        f"The pressure in section 2 is {pressure_2}, what is the pressure in section 3?",
        space=9,
    ),
]

with open("4-pipo.png", "rb") as f:
    encoded_string = base64.b64encode(f.read())

wp = WorkProblem(statement=statement, questions=questions, figure=encoded_string)
problems.append(wp)
if mode == "dev":
    wp.show(answers=True)
```

```python
mass_bridge = Value(500, "kg")
mass_object = Value(25, "kg")
angle = Value(37, "degrees")
break_force = Value(98.9, "kN")

statement = f"In the following figure, the bridge mass is {mass_bridge}, the hanging object is {mass_object}. The rope is attached to the bridge $\\frac{{4}}{{10}}$ of the distance from the hinge, and the angle shown is $\\phi={angle:C}$."

with open("bridge.png", "rb") as f:
    encoded_string = base64.b64encode(f.read())

questions = [
    Question(
        "Draw out all the forces acting on the bridge, rope, and hinge.",
        space=15,
        figure=encoded_string,
    ),
    Question(
        "Write down the equations needed to keep the bridge in equilibrium.", space=9
    ),
    Question("Solve for the tension in the rope.", space=9),
    Question(
        f"What would the mass of the object need to be to snap the rope (it is made of a one inch nylon rope that has a breaking strength of {break_force})?",
        space=9,
    ),
]


wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
if mode == "dev":
    wp.show(answers=True)
```
