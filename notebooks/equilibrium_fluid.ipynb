```python
from testifyr import WorkProblem, Question, Test, MultipleChoiceProblem, Choice, Bonus, Information
from stuff import Value, Unit
import numpy as np
from pprint import pprint as pp
import base64

problems = []
multiple_choices = []
bonuses = []
g = Value(9.81, 'm/s^2')
mode = "dev1"
```

```python

mc = MultipleChoiceProblem(statement="What is the relationship between mass, volume, and density?", choices=[
    Choice("Density is inversely proportional to both mass and volume."),
    Choice("Density is proportional to both mass and volume."),
    Choice("Density is proportional to mass and inversely proportional to volume.", True),
    Choice("Density is inversely proportional to mass and proportional to volume."),
])
multiple_choices.append(mc)
if mode == "dev": mc.show()
```

```python
depth_a = Value(2, "in")
depth_b = Value(1.5, "ft")

mc = MultipleChoiceProblem(statement=f"Zaphod is holding two copies of the Hitchhiker's Guide to the Galaxy (same volume, same density) in a pool of liquid gallium. One of the guides (A) is held {depth_a} below the surface and the other (B) is held at a depth of {depth_b}. Which one experiences a greater buoyant force?", choices=[
    Choice("Both feel the same force.", True),
    Choice("Guide A."),
    Choice("Guide B."),
    Choice("Need more information.")
])
multiple_choices.append(mc)
if mode == "dev": mc.show()
# mc.html()
```

```python
rho_water = Value(1000, "kg/m^3")
rho_gallium = Value(6095, "kg/m^3")

def rho(var="", v=None):
    return f"$\\rho_\\mathrm{{{var}}} = {v:C}$"
print(rho('w', rho_water))

mc = MultipleChoiceProblem(statement=f"Now A is held {depth_a} below the surface in liquid gallium ({rho('gallium', rho_gallium)}) and B is held {depth_b} under water ({rho('water', rho_water)}), which one experiences a greater buoyant force?", choices=[
    Choice("Both feel the same force."),
    Choice("Guide A.", True),
    Choice("Guide B."),
    Choice("Need more information.")
])
multiple_choices.append(mc)
if mode == "dev": mc.show()
```

```python
rho_guide = Value(2183, "kg/m^3")

mc = MultipleChoiceProblem(statement=f"In the case of Guide B (({rho('Guide', rho_guide)}), $\\mathrm{{Depth}}={depth_b:C}$), will it float up in either fluid?", choices=[
    Choice("Both."),
    Choice("Only gallium.", True),
    Choice("Only water."),
    Choice("Neither")
])
multiple_choices.append(mc)
if mode == "dev": mc.show()
```

```python
bonus = Bonus(statement="Define incompressible in an incomprehensible way.", answer="Noffer trunions are operated by a turbo-encabulator. ")
multiple_choices.append(bonus)
if mode == "dev": bonus.show()
```

```python
statement = "Which of the following tells us how to relate the density of an object with it's volume displacement and force of buoyancy?"
choices = [
    Choice("Pascal's principle."),
    Choice("Continuity equation."),
    Choice("Bernoulli's equation."),
    Choice("Newton's law."),
    Choice("Archimede's principle.", True)
]
# with open('door.png', 'rb') as f:
#     encoded_string = base64.b64encode(f.read())
mc = MultipleChoiceProblem(statement=statement, choices=choices) #, figure=encoded_string)
multiple_choices.append(mc)
#mc.show()
```

```python
stool_omega = Value(10, 'rad/s')
statement=f"A person is sitting on a spinning stool rotating at {stool_omega}. They are holding weights close to their chest. The person then extends their arms outwards. What happens to the stool's angular velocity?"
choices = [
    Choice("It increases"),
    Choice("It decreases", True),
    Choice("It stays the same"),
    Choice("It depends on the mass of the weights")
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
#mc.show()
```

```python
statement = "The wrench in the figure has six forces of equal magnitude acting on it."
with open('wrench.jpg', 'rb') as f:
    encoded_string = base64.b64encode(f.read())
questions = [
    Question("Rank these forces (A through F) on the basis of the magnitude (greatest to least) of the torque they apply to the wrench, measured about an axis centered on the bolt. If two are the same, list them alphabetically", figure=encoded_string, space=10, answer="D > B > E > F > A > C"),
]
wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
#wp.show()
```

```python
mass_dvd = Value(16, 'g')
diameter_dvd = Value(12, 'cm')
angular_speed_dvd = Value(9280, 'rpm', sigfigs=4)
time_stop = Value(2.5, 's')
close_distance = Value(2, 'cm')

statement = f"A DVD (disc) has a mass of {mass_dvd} and a diameter of {diameter_dvd}. The player is spinning it at an angular speed of {angular_speed_dvd} (one rotation is $2\\pi$ radians) when it is turned off and slows to a stop in {time_stop}."

questions = [
    Question("What is the moment of inertia of the DVD?", space=9),
    Question("What is the magnitude and direction of the torque felt by the DVD?", space=9),
    Question("What is the initial angular momentum of the DVD?", space=9),
    Question(f"What is the linear (straight line) speed of two points on the DVD: a point at the edge of the DVD, and a point {close_distance} from the center?", space=9),
]
wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
#wp.show(answers=True)
```

```python
bonus = Bonus(statement="Evaluate the definite integral $\\int_0^{2} (x+3x^2) dx$. (recall: integral means anti-derivative, or area under a curve)", answer="10")
problems.append(bonus)
#bonus.show()
```

```python
diameter_ball = Value(0.314, 'm')
height_initial = Value(325, 'm')
speed_initial = Value(5, 'm/s')

statement = f"A rubber ball (solid sphere) with a diameter of {diameter_ball} is rolling without slipping with an initial linear speed $v={speed_initial:C}$ from the top of a hill, {height_initial} high, to the bottom."

questions = [
    Question("Draw a diagram showing this situation. Include a before and after picture labeling all relevant quantities.", space=15),
    Question("Derive an equation for the total kinetic energy in terms of $m$ and $v$.", space=9),
    Question("When the ball reaches the bottom of the hill what will its linear speed be?", space=9),
    Question("What will be the rotational speed of the ball at the bottom of the hill?", space=9),
]
wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
#wp.show(answers=True)
```

```python
test = Test({}, multiple_choices + problems)
# print(test.total_points())
# print('\n'.join(test.get_grade_list()))
```
