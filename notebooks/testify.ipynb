```python
from testifyr import WorkProblem, Question, Test, MultipleChoiceProblem, Choice
from stuff import Value, Unit
problems = []
multiple_choices = []
```

```python
v_i = Value(26.8, "meter / second")
theta_i = Value(18.3, "degree")
x_f = Value(30.5, "meter")
mass = Value(0.43, "kilogram")

statement = fr"Joseph Fourier tosses a stone of mass of {mass} with a speed of {v_i} at an angle of {theta_i} degrees above the horizontal towards a wall that is {x_f} away."

questions = [
    Question("Create and draw a well labeled diagram of the situation. Be sure to include all known and unknown quantities.", 10),
    Question("Resolve the initial velocity into $x$ and $y$ components:", 5),
    Question("What is the time ($t$) taken for the stone to hit the wall?", 5),
    Question("What is the height of the stone when it hits the wall?", 5),
]

wp = WorkProblem(statement, questions)
problems.append(wp)
#wp.show()
```

```python
Unit.add_unit('blark = 0.592 yard')
one_blark = Value(1, 'blark')

Unit.add_unit('zoomer = 1/0.833 hour')
one_zoomer = Value(1, 'zoomer')

a = Value(2.45, "blark / zoomer")

statement = fr"Quinn accelerates at a rate of {a}."
questions = [
    Question(fr"Determine Quinn's acceleration in ${Unit('m/s'):~L}$.", 3),
]
extras = [
    fr"{one_blark} $=$ {one_blark.to('yard')}",
    fr"{one_zoomer} $=$ {one_zoomer.to('hour')}",
]

wp = WorkProblem(statement, questions)#, extras=extras))
problems.append(wp)

#wp.show()
```

```python
mass = Value(97.1, "kg")
x = Value(36.6, "m")
v_f = Value(8.02, "m/s")

statement = fr"Alvin Kamara has a mass of {mass} and ran {x}, accelerating to a speed of {v_f}."
questions = [
    Question(fr"Determine Alvin's acceleration.", 3),
    Question(fr"Determine Alvin's time in the {x} dash.", 3),
]

wp = WorkProblem(statement, questions)
problems.append(wp)
#wp.show()
```

```python
sig_figs = 3
import math
number = Value(math.pi, pre_zeros=3, sigfigs=sig_figs)
statement = f"How many significant figures are in the number {number}?"
choices = [Choice(i, sig_figs == i) for i in range(2, 6)]
mc = MultipleChoiceProblem(statement, choices)
multiple_choices.append(mc)
#mc.show()
```

```python
statement = "For general projectile motion, which of the following best describes the horizontal and vertical components of a projectile's acceleration? (assume air resistances is negligible):"
choices = [
    Choice("$a_x = 0$, $a_y = -g$", True),
    Choice("$a_x = -g$, $a_y = 0$"),
    Choice("$a_x = 0$,  $a_y = g$"),
    Choice("$a_x = g$,  $a_y = 0$"),
]
mc = MultipleChoiceProblem(statement, choices)
multiple_choices.append(mc)
#mc.show()
```

```python
statement = "A red water balloon is thrown horizontally from the top of a bridge. At the same instant, a yellow water balloon is dropped off the bridge from the same height. Compare the time of fall for the two balloons. Assume air resistance is negligible."
choices = [
    Choice("The red balloon will hit the ground first."),
    Choice("The yellow balloon will hit the ground first."),
    Choice("Both balloons will hit the ground at the same time.", True),
    Choice("It is impossible to determine without more information."),
]
mc = MultipleChoiceProblem(statement, choices)
multiple_choices.append(mc)
#mc.show()
```

```python
test = Test({}, multiple_choices + problems)
test.show()
```
