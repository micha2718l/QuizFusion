```python
from testifyr import WorkProblem, Question, Test, MultipleChoiceProblem, Choice, Bonus, Information
from stuff import Value, Unit
import numpy as np
from pprint import pprint as pp
import base64

problems = []
multiple_choices = []
bonuses = []
g = Value(9.81, 'm/s^2')
```

```python
middle_angle = Value(45, 'degree')

info = Information(f"For the following 3 problems, we are referring to a situation where a box is sliding down various ramps. The ramps are listed as A, B, and C. A has the steepest angle and C has the least steep angle. B is in the middle, something like a {middle_angle} angle. Assume no friction.")
multiple_choices.append(info)
mc = MultipleChoiceProblem(statement="With which ramp will the box reach the bottom fastest?", choices=[
    Choice("A"),
    Choice("B"),
    Choice("C"),
    Choice("They will all take the same amount of time.", True),
    Choice("It is impossible to determine without more information.")
])
multiple_choices.append(mc)
mc = MultipleChoiceProblem(statement="With which ramp will the box reach the bottom with the greatest speed?", choices=[
    Choice("A"),
    Choice("B"),
    Choice("C"),
    Choice("They will all reach the bottom with the same speed.", True),
    Choice("It is impossible to determine without more information.")
])
multiple_choices.append(mc)
mc = MultipleChoiceProblem(statement="Will all of the boxes have the same velocity when they reach the bottom?", choices=[
    Choice("Yes"),
    Choice("No", True)
])
multiple_choices.append(mc)
```

```python
statement = "Which of the following types of collisions has the quality that both kinetic energy and momentum are conserved?"
choices = [
    Choice("Inelastic"),
    Choice("Elastic", True),
    Choice("Perfectly inelastic"),
    Choice("None of the above")
]

multiple_choices.append(MultipleChoiceProblem(statement=statement, choices=choices))
```

```python
g_risa = Value(1.01* 9.81, 'm/s^2')
mass_riker = Value(90.7, 'kg')
height_riker = Value(188, 'cm')
height_initial = Value(42, 'm')
length_rope = Value(27, 'm')
mass_latinum = Value(17, 'kg')

final_speed = Value(2.5, 'm/s')

statement = f"For the following 3 problems, we are referring to the following situation: Commander William T. Riker (height {height_riker} and mass {mass_riker}) is on vacation on planet Risa (gravitational acceleration {g_risa}) and comes across a river he wants to cross to get to the Risians on the other side having a party. He must swing across with a {length_rope} long rope from an initial height of {height_initial} and he knows that he must let go at the right time to land on the other side, but he doesn't know when or where. He does know that he must let go when his speed is exactly {final_speed}. Assume no friction or air resistance."
info = Information(statement)
multiple_choices.append(info)
# info.show()
statement = f"Which of the following pieces of information is not needed to determine when and where Riker should let go of the rope?"
choices = [
    Choice("The gravitational acceleration of Risa"),
    Choice("The mass of Riker", True),
    Choice("The final speed of Riker"),
    Choice("The initial height swung from"),
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
# mc.show(answers=True)
height_equation = r"$h_f=\frac{g h_{i} - \frac{1}{2} v}{g}$"
statement = f"The commander (incorrectly) calculates that he must let go of the rope when his height is determined by the following (wrong) equation: {height_equation}. Which of these tells us this is wrong?"
choices = [
    Choice("The equation is not dimensionally consistent", True),
    Choice("He forgot to include his mass"),
    Choice("The equation is missing a minus sign"),
    Choice("The equation has one too many minus signs"),
    Choice("The $\\frac{1}{2}$ should be a $2$"),
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
# mc.show(answers=True)
statement = f"There is a Ferengi chasing him from the starting side of the river demanding to be paid for a game of Dabo. What is the result of Riker throwing {mass_latinum} of latinum to the Ferengi at the moment he lets go?"
choices = [
    Choice("Riker will not launch at all (drop straight down)"),
    Choice("Riker will launch with the same speed"),
    Choice("Riker will launch with a greater speed", True),
    Choice("Riker will launch with a lesser speed"),
    Choice("It is impossible to determine without more information.")
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
# mc.show(answers=True)
```

```python
mass_ball = Value(0.75, 'kg')
height = Value(5, 'm')
speed = Value(10, 'm/s')

statement = f"A {mass_ball} ball drops from a height of {height} meters. What is the speed of the ball just before it hits the ground?"
choices = [
    Choice(f"{speed/2}"),
    Choice(f"{speed}", True),
    Choice(f"{speed*2}"),
    Choice("Not enough information to determine")
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
# mc.show(answers=True)
```

```python
mass_box = Value(5.0, 'kg')
spring_constant = Value(350, 'N/m')
spring_compression = Value(0.2, 'm')
energy_dissipated = Value(2.718, 'J')
slide_distance = Value(2.5, 'm')

statement = f"A box of mass {mass_box} is pushed against a spring with spring constant {spring_constant} until the spring is compressed by {spring_compression}. The box is then released and slides across a rough surface. The box comes to a stop due to friction after sliding {slide_distance}."

questions = [
    Question("Draw a diagram showing this situation. Be sure to include a before and after picture and label all relevant quantities.", space=20),
    Question("What is the maximum speed of the box?", space=10),
    Question("What is the force due to friction and how much total work is done to stop the box?", space=10)
]
wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
#wp.show(answers=True)
bonus = Bonus("What is the coefficient of kinetic friction between the box and the surface?")
problems.append(bonus)
#bonus.show(answers=True)
```

```python
mass_cannon = Value(1234, 'kg')
speed_cannonball = Value(125, 'm/s')
speed_cannon = Value(5, 'm/s')

statement = f"A cannon of mass {mass_cannon} is loaded with a heavy cannonball and fired horizontally to the right. The cannonball is fired with a speed of {speed_cannonball} and the cannon recoils with a speed of {speed_cannon}. The cannon is on a frictionless surface, and air resistance is negligible."

questions = [
    Question("Draw a diagram showing this situation. Include a before and after picture labeling all relevant quantities.", space=20),
    Question("Write down the relevant momentum conservation equation. Solve for the mass of the ball (no numbers).", space=10),
    Question("What is the mass of the cannonball?", space=10)
]
wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
#wp.show(answers=True)
```

```python
bonus = Bonus(f"How many times must one say Beetlejuice's name to use the same energy as Betelgeuse emits in one second?")
problems.append(bonus)
#bonus.show(answers=True)
#bonus.html()
```
