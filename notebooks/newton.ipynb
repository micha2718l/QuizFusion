```python
mass_truck = Value(12.3, 'Mg')
mass_bug = Value(0.4321, 'g')

statement = f"A truck with mass {mass_truck} collides with a bug with mass {mass_bug}. How does the magnitude of the bug's force on the truck compare to the magnitude of the truck's force on the bug?"

choices = [
    Choice("The bug's force on the truck is greater than the truck's force on the bug."),
    Choice("The truck's force on the bug is greater than the bug's force on the truck."),
    Choice("The two forces are equal in magnitude.", True),
    Choice("It is impossible to determine without more information.")
]

mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
#mc.show(answers=True)
```

```python
mass_box = Value(10, 'kg')

statement = f"A {mass_box} box is at rest on an inclined ramp. Which of the following correctly describes the magnitude of force due to static friction?"

choices = [
    Choice(r"$|\vec{f_s}|=|\vec{W}|$"),
    Choice(r"$|\vec{f_s}|=|\vec{N}|$"),
    Choice(r"$|\vec{f_s}|=|\vec{W_\bot}|$"),
    Choice(r"$|\vec{f_s}|=|\vec{W_\parallel}|$", True),
    Choice(r"$|\vec{f_s}|=0$"),
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
#mc.show(answers=True)
```

```python
mass_arthur = Value(70, 'kg')
acceleration = Value(2, 'm/s^2')
g_sqornshellous = Value(5.7, 'm/s^2')

statement = fr"Arthur Dent is in an elevator on the planet Squornshellous Zeta where the gravitational acceleration is $g_{{\mathrm{{Squornshellous}}}}={g_sqornshellous:C}$. The elevator is accelerating upwards at {acceleration}. What is the magnitude of the normal force on Arthur if his mass is {mass_arthur}?"
choices = [
    Choice(fr"$N = m g = {mass_arthur * g_sqornshellous:C}$"),
    Choice(fr"$N = m g + m a = {mass_arthur*(g_sqornshellous + acceleration):C}$", True),
    Choice(fr"$N = m g - m a = {mass_arthur*(g_sqornshellous - acceleration):C}$"),
    Choice(fr"$N = m a = {mass_arthur * acceleration:C}$"),
]
mc = MultipleChoiceProblem(statement=statement, choices=choices)
multiple_choices.append(mc)
#mc.show(answers=True)
```

```python
mass_godric = Value(9.98, 'kg')
acceleration_elevator = Value(2, 'm/s^2')

statement = fr"Baby Godric (who currently has a mass of {mass_godric}) is being held up by his mom as in the picture. Treat each of his arms as ideal tension ropes."

fn = "godricHanging.png"
with open(fn, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read())

with open('problem9question1_godricFreeBody_Algier.jpg', "rb") as image_file:
        box_pulley_encoded_string = base64.b64encode(image_file.read())
box_pulley_answer_diagram = f"<img style='height:100%;' src='data:image/jpg;base64,{box_pulley_encoded_string.decode()}' />"

questions = [
    Question("Draw a free body diagram for Godric showing all forces acting on him.", space=20, figure=encoded_string, answer=box_pulley_answer_diagram),
    Question("What is the tension in his right arm?", space=10,
             answer="\n\n".join([fr"$T_\mathrm{{L}} + T_\mathrm{{R}} = m g = {mass_godric:C} * {g:C}$",
                fr"$T_\mathrm{{L}} = {mass_godric * g:C} - T_\mathrm{{R}}$",
                fr"Notice: $T_\mathrm{{L}} = T_\mathrm{{R}}$",
                fr"$T_\mathrm{{L}} = T_\mathrm{{R}} = \frac{{m g}}{{2}} = {mass_godric * g / 2:C}$"
                ])
             ),
    Question(f"What would the tension in the right arm be if they were in an elevator accelerating upward at {acceleration_elevator}?", space=10,
             answer="\n\n".join([fr"$T_\mathrm{{L}} + T_\mathrm{{R}} = m (g + a) = {mass_godric:C} * ({g:C} + {acceleration_elevator:C})$",
                fr"$T_\mathrm{{L}} = {mass_godric * (g + acceleration_elevator):C} - T_\mathrm{{R}}$",
                fr"Notice: $T_\mathrm{{L}} = T_\mathrm{{R}}$",
                fr"$T_\mathrm{{L}} = T_\mathrm{{R}} = \frac{{m (g + a)}}{{2}} = {mass_godric * (g + acceleration_elevator) / 2:C}$"
                ])
             ),
]
wp = WorkProblem(statement=statement, questions=questions)
g_mars = Value(3.7, 'm/s^2')
bonus = Bonus("What would the tension in the left arm be if they were on the surface of Mars?", 3,
              answer="\n".join([f"Gravity on Mars is {g_mars}. (from test instructions), ",
                f"$T_1 = T_2 = \\frac{{m g_\\mathrm{{mars}}}}{{2}} = {mass_godric * g_mars / 2:C}$"
                ])
             )
problems.append(wp)
problems.append(bonus)
#wp.show(answers=True)
#bonus.show(answers=True)
```

```python
#with open('boxes_pulley.svg', 'r') as f:
#    boxes_pulley_svg = f.read()
with open('boxes_pulley.png', 'rb') as f:
    encoded_string = base64.b64encode(f.read())

mass_box1 = Value(10, 'kg')
mass_box2 = Value(5, 'kg')
friction_coefficient_kinetic = Value(0.2, 'dimensionless')
statement = fr"Two boxes are connected by a string (assume massless, frictionless string that never breaks) over a pulley. Box 1 has mass {mass_box1} and is on a table where the coefficient of kinetic friction is $\mu_k={friction_coefficient_kinetic:C}$. Box 2 has mass {mass_box2} and is hanging from the string. The system is in motion (meaning $v_{{x_{{\mathrm{{Box1}}}}}} \ne 0$)."

a = ((mass_box2 * g) - (friction_coefficient_kinetic * mass_box1 * g)) / (mass_box1 + mass_box2)
with open('problem11question1_figure_Tao.jpg', "rb") as image_file:
        box_pulley_encoded_string = base64.b64encode(image_file.read())
box_pulley_answer_diagram = f"<img style='height:100%;' src='data:image/jpg;base64,{box_pulley_encoded_string.decode()}' />"

questions = [
    Question("Draw a free body diagram for each box showing all forces acting on it.", space=14, figure=encoded_string, answer=box_pulley_answer_diagram),
    Question("What is the acceleration of the system? (only give algebraic answer, no numbers yet)", space=9,
             answer="\n\n".join([fr"$\sum_{{\mathrm{{ext}}}}^{{}}{{F_x}} = m_1 a_x = T_1 - f_k$ and $\sum_{{\mathrm{{ext}}}}^{{}}{{F_y}} = m_2 a_y = m_2 g - T_2$",
                    fr"$a_x = a_y = a$ and $T_1 = T_2 \implies T = f_k + m_1a = T = m_2 g - m_2 a$",
                    fr"$m_1 a + m_2 a = m_2 g - f_k \implies a = \frac{{m_2 g - f_k}}{{m_1 + m_2}}$",
                    "Notice this implies the acceleration is positive when the block on the table is moving to the left."
                ])
    ),
    Question("What is the acceleration of the system? (give a number with units, 3 significant figures)", space=9,
             answer="\n\n".join([fr"$a = \frac{{m_2 g - f_k}}{{m_1 + m_2}}$, $f_k=\mu_k N$, $N=m_1 g$, $f_k=\mu_k m_1 g$",
                    fr"$a = \frac{{ ({mass_box2:C}) ({g:C}) - ({friction_coefficient_kinetic:C}) ({mass_box1:C}) ({g:C})}}{{{mass_box1:C} + {mass_box2:C}}}$",
                    fr"$a = {a:C}$"
                ])
             ),
    Question("What is the tension in the string?", space=9,
             answer="\n\n".join([fr"$T = f_k + m_1 a = \mu_k m_1 g + m_1 a = ({friction_coefficient_kinetic:C}) ({mass_box1:C}) ({g:C}) + ({mass_box1:C}) ({a:C})$",
                    fr"$T={friction_coefficient_kinetic * mass_box1 * g + mass_box1 * a:C}$"
                                 
                                 #T = m_2 g - m_2 a"
                ])
             ),
]
wp = WorkProblem(statement=statement, questions=questions)
problems.append(wp)
#wp.show(answers=True)
```
